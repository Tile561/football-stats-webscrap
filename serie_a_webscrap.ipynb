```python
from bs4 import BeautifulSoup as bs, Comment
import requests
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python

url = "https://fbref.com/en/comps/11/stats/Serie-A-Stats"
headers = {'User-Agent': 'Mozilla/5.0'} 
response = requests.get(url, headers=headers)
```

```python
soup = bs(response.text)
type(soup)
soup.find('title').text
```

```python
#extracting the headers
seria_table = soup.find('table', id="stats_squads_standard_for")
seria_thead = seria_table.find('thead')

seria_header = seria_thead.find_all('th')
#print(header)
columns = [ele.text.strip() for ele in seria_header]
#print(columns)


#creating subheaders and headers
n = 6 
over_header = columns[:n]
under_header = columns[n:]
over_header_replaced = over_header[0]="Empty"

#structure of the subheader and header
structure = {
    'Empty': 4,
    'Playing Time': 4,
    'Performance': 8,
    'Expected': 4,
    'Progression': 2,
    'Per 90 Minutes': 10
}

over_headers =[]
for key, count in structure.items():
    over_headers.extend([key]*count)

#creating the subheader and header
multi_index = pd.MultiIndex.from_arrays([over_headers, under_header])
```

```python
# extracting the rows
seria_tbody = seria_table.find('tbody')


rows = seria_tbody.find_all('tr')
d=[]
for row in rows:
    columns = row.find_all(['td', 'th'])
    row = [ele.text.strip() for ele in columns]
    d.append((row))
    

seria_squad_df = pd.DataFrame(d, columns=multi_index)
```
